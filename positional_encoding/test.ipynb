```python
import numpy as np
```

```python
d_model = 6
seq_length = 10
```

```python
pos = np.arange(seq_length)[:, np.newaxis]
print(pos)
pos.shape
```

```python
i = np.arange(d_model)[np.newaxis, :]
print(i)
i.shape
```

```python
angle_rates = np.power(10000, (2 * (i//2)) / d_model)
print(angle_rates)
angle_rates.shape
```

```python
angles = angle_rates * pos
print(angles)
angles.shape
```
